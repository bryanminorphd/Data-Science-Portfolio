The above graph shows that the European countries (red dots) had lower birth rate and higher life expectancy in the 60's. On the otherhand, African countries had significantly higher birth rate and lower life expectancy. We can say that life expectancy in Africa ranged from 30-50 years.

Comparison of the two charts: 

Looking at the abive charts, we see that

1. The overall life expectancy in the world has increased since 1960.

3. African countries still maintain higher birth rate with few exceotions. Life expectancy has somewhat increased.
4. Life expectancy has gone up in the Asian and Middle Eastern countries, even though some countries still have higher birth rate.
5. Countries in America and Oceania follows somewhat similar pattern as the 60's.

It can be concluded that higher birth rate negatively affects life expectancy.
